```python
import pandas as pd
```

```python
df = pd.read_csv('filtered_trades.csv')
```

```python
for ticker in df['ticker']:
    data = pd.read_csv('Stocks/' + ticker + '.us.txt')
    data.drop(['Open', 'High', 'Low', 'Volume', 'OpenInt'], axis = 1,inplace=True)
    data.to_csv('data/' + ticker + '.csv')
```
